```python
distance_in_km_car = [62, 67, 72, 74, 69]
time_to_cover_in_min = [155, 220, 240, 195, 175]


distance_in_km_bike = [55,57, 82, 78, 86]
time_to_cover__by_bike_in_min = [145, 210, 250, 198, 165]

plt.title("Distance vs Time relationsip of car")
plt.xlabel("distance in km to cover by car")
plt.ylabel("time to cover distance in min")

plt.scatter(distance_in_km_car, time_to_cover_in_min ,label = "Car")  
plt.scatter(distance_in_km_bike, time_to_cover__by_bike_in_min, label ="Bike")

plt.legend()
```
